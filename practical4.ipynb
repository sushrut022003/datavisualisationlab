```python
# Dealing with Missing Values 
import pandas as pd
import numpy as np
df=pd.read_csv('Employee.csv')
df
```

```python
no_of_emp=0
sum=0
for cgpa in df['GradCGPA']:
    if str(cgpa).replace('.','').isdigit():
        no_of_emp+=1
        sum+=cgpa
average=sum/no_of_emp
df=df.replace(to_replace=np.nan,value=average)
df
```

```python
# Data Reshaping
# mapping gender column male value with 0 and female with 1
df['Gender']=df['Gender'].map(
    {'Male':0,'Female':1}
)
df
```

```python
# Filtering data
df=df[df['GradCGPA']>8.0]
df
```

```python
# Merging Data
import pandas as pd
student={
    'RollNo' : ['17','09','13','10','12','06'],
    'Name':['Sam','Gauy','Tejaml','Aashim','Morer','Jash']
}
project={
     'RollNo' : ['17','09','13','10','12','06'],
     'Task':['Opencv','NLP','DeepLearning','Website','Database','FlaskFramework']
}
student_data=pd.DataFrame(student)
project_data=pd.DataFrame(project)
```

```python
print(student_data)
```

```python
print(project_data)
```

```python
# merging this two data with common column roll no
df=pd.merge(student_data,project_data,on='RollNo')
df
```
